```python
entropy_at_step = []
for i in range(0,df.shape[0]):
    row = df.iloc[i,1:]
    prob_x=0 
    unique_value_counts = row.value_counts()
    for value, count in unique_value_counts.items():
        prob_x = prob_x - (count/400) * (math.log2(count/400))
    entropy_at_step.append(prob_x)

plt.figure(figsize=(10, 6))
plt.xlabel('Step In Time')
plt.ylabel('Shannon Entropy')
plt.title('Entire Black Box')
x_values=list(range(0,2000))
y_values=entropy_at_step[0:2000]
plt.plot(x_values, y_values, color='purple',label='Y Values')
plt.show()
```
